# Text preprocessing
- Clean Text
    - remove all special character in the text
    - covert all text into lower case
- Tokenization
    - get only words or tokens
    - convert string into list of words
- Stemming
    - remove all morphological affixes from word like past, continous, prual
- Lemma
    - take the root word

```python
def clean_text(text):
    text = text.lower() # convert into lowercase
    text = re.sub(r'[^a-z0-9 ]','',text) # remove special characters
    return text
```

```python
text
```

```python
clean_txt = clean_text(text)
clean_txt
```

#### Tokenization

```python
#clean_txt.split(' ')
tokens = nltk.word_tokenize(clean_txt)
print(tokens)
```

### Stemming

```python
word = 'caring'
snow = nltk.stem.SnowballStemmer(language='english')
```

```python
snow.stem('cared')
```

```python
token_stem =[]
for w in tokens:
    w = snow.stem(w) # stem
    token_stem.append(w)
```

```python
print(token_stem)
```

```python
" ".join(token_stem) # joining all list of word
```

### Lemma
- it will give the root word and also remove past, continous etc
- it is slow

```python
nltk.download('wordnet')
```

```python
lemma = nltk.wordnet.WordNetLemmatizer()
```

```python
lemma.lemmatize('caring',pos='v')
```

Parts of Speech:
1. noun, 2. pronoun, 3. Adjective 4. Verb 5. Adverb 6. Conjunction 7. preposition 8.Interjection

```python
tagging  = nltk.pos_tag(tokens)
print(tagging)
```

```python
lemma.lemmatize('are',pos='v') # n -noun, v - adverb, a- adjective, r - adverb
```

only "verb", "adjective", "Adverb"

pos = 'PRP'
pos.startswith('P')

```python
tok_lemma = []
for word, pos in tagging:
    if pos.startswith('V'):
        tag = 'v'
    elif pos.startswith('R'):
        tag = 'r'
    elif pos.startswith('J'):
        tag ='a'
    else:
        tag ='n'
    
    # lemma 
    tok_lem = lemma.lemmatize(word,pos=tag) 
    tok_lemma.append(tok_lem)
```
